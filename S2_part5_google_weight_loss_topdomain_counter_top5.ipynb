```python
# Most common websites in RANK_01
s1 = dataset.groupby('Rank_01').size()
df1 = s1.to_frame(name='Count')
df1 = df1.sort_values(by='Count', ascending=False, na_position='first')
df1[:10]
```

```python
# Most common websites in RANK_02
s2 = dataset.groupby('Rank_02').size()
df2 = s2.to_frame(name='Count')
df2 = df2.sort_values(by='Count', ascending=False, na_position='first')
df2[:10]
```

```python
# Most common websites in RANK_03
s3 = dataset.groupby('Rank_03').size()
df3 = s3.to_frame(name='Count')
df3 = df3.sort_values(by='Count', ascending=False, na_position='first')
df3[:10]
```

```python
# Most common websites in RANK_04
s4 = dataset.groupby('Rank_04').size()
df4 = s4.to_frame(name='Count')
df4 = df4.sort_values(by='Count', ascending=False, na_position='first')
df4[:10]
```

```python
# Most common websites in RANK_05
s5 = dataset.groupby('Rank_05').size()
df5 = s5.to_frame(name='Count')
df5 = df5.sort_values(by='Count', ascending=False, na_position='first')
df5[:10]
```

```python
# Transformind dataframes into dictionaries to later merge
dic1 = dict(df1.iloc[:,-1])
dic2 = dict(df2.iloc[:,-1])
dic3 = dict(df3.iloc[:,-1])
dic4 = dict(df4.iloc[:,-1])
dic5 = dict(df5.iloc[:,-1])
```

```python
# Merging dic1 and dic2
result_1 = {key: dic1.get(key,0) + dic2.get(key,0) for key in set(dic1) | set(dic2)}

# Merging result_1 and dic3
result_2 = {key: result_1.get(key,0) + dic3.get(key,0) for key in set(result_1) | set(dic3)}

# Merging result_2 and dic4
result_3 = {key: result_2.get(key,0) + dic4.get(key,0) for key in set(result_2) | set(dic4)}

# Merging result_3 and dic5
result_4 = {key: result_3.get(key,0) + dic5.get(key,0) for key in set(result_3) | set(dic5)}
```

```python
# Creating dataframe from merged dictionaries
new = pd.DataFrame(result_4, index=['Total'])
new = new.transpose()
```

```python
# Write dataframe into csv file
with open('google_weight_loss_TOTAL_top_5_final.csv', 'w') as w:
    new.to_csv(w)
    w.close
```
