```python
import os
import pandas as pd
from urllib.parse import urlparse
```

```python
# Set working directory
os.chdir('/home/user/Documents')
```

```python
# Loading the dataset of 432 queries and their respective link results 
with open('google_weight_loss_results.csv', 'r', encoding='utf-8') as f:
    dataset = pd.read_csv(f, index_col=0)
    f.close
dataset.head()
```

```python
# Subsetting the first 5 results
df = dataset.iloc[:5]
df.head()
```

```python
# A list of our queries
queries = df.columns.tolist()
print(queries[:5])
```

```python
# Replacing URLs with top level domain
for i in queries:
    for l in df[i]:
        sub_domain = urlparse(l).netloc
        top_domain = '.'.join(sub_domain.split('.')[-3:])
        df[i] = df[i].replace(l, top_domain)
# NOTE: there are some caveats of using this method, see: https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#returning-a-view-versus-a-copy
df.head()
```

```python
# Replace www.
df2 = df.replace('www.','', regex=True)
df2.head()
```

```python
# Is there any missing value?
nan = df2.isnull().values.any()
print(nan)
```

```python
# Write dataframe into csv file
with open('google_weight_loss_results_topdomain_new_test.csv', 'w') as w:
    df2.to_csv(w)
    w.close
```
