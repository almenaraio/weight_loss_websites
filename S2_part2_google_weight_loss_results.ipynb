```python
import os
import time
import pandas as pd
import json
```

```python
# Set working directory
os.chdir('/home/user/google_weight_loss/results')
```

```python
# Make a list of files in folder WITH file extension
path = '/home/user/google_weight_loss/results' 
list1 = os.listdir(path)
print(list1[:10])
```

```python
# Make a list of files in folder WITH and WITHOUT file extension
list2 = [os.path.splitext(filename)[0] for filename in os.listdir(path)]
print(list2[:10])
```

```python
# Let's create a list of lists for all link results
data = []
start = time.time()
for i in list1:
    with open(i, 'rb') as f:
        query = json.load(f)
        results = query['organic_results']
        links = [ val['link'] for val in results ]
        data.append(links)
        f.close
end = time.time()
print('Finished after ' + str(round(end - start, 2)) + ' seconds.')
```

```python
# Creating a dataframe: columns are queries, rows are link results
df = pd.DataFrame(data)
df = df.transpose()
df.columns = list2
df.head()
```

```python
# Write dataframe into csv file
with open('google_weight_loss_results.csv', 'w') as w:
    df.to_csv(w)
    w.close
```
